```python
%%R
set.seed(0)
instance = function(n, p, g, max.steps=5, S=1.5) {
X = matrix(rnorm(n*p), n, p)
Y = rnorm(n) * S
groups = rep(1:g,(p/g)+1)[1:p]
Sigma = S^2 * diag(rep(1,n)) 
return(forward_group(X, Y, groups, Sigma=Sigma, max.steps=max.steps)$p.vals)
q}

instance(53,121,23)
```

```python
%%R
nsim = 500
set.seed(0)
P = c()
for (i in 1:nsim) {
    P = rbind(P, instance(53,123,23))
}
```

```python
%%R -h 800 -w 800
par(mfrow=c(2,2))
for (i in 1:4) {
plot(ecdf(P[,i]))
abline(0,1)
}
```
